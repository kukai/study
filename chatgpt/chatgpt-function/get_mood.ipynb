```python
import openai
import json
import os

openai.api_key = os.getenv("OPENAI_API_KEY")

def get_mood(text):
    response = openai.Completion.create(
      engine="text-davinci-002",
      prompt=(f"Quantify the mood of the following text:\n{text}\nMood (on a scale of -100 to 100):"),
      temperature=0.5,
      max_tokens=100,
      n=1,
      stop=None,
      timeout=15,
    )
    # print(response)
    mood_score = int(response.choices[0].text.strip())

    
    output = {
        "text": text,
        "mood_score": mood_score
    }
    return json.dumps(output)

text = "今朝はよく眠れました。明日は休日なので遊びに行きます。"
output = get_mood(text)
print(output)
```
