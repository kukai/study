```python
import random
import gradio as gr

def get_foturne(your_name):
    fortune_lists = ['大吉', '吉', '小吉', '凶', '末吉']
    fortune_result = random.choice(fortune_lists)
    return your_name + "さんの今日の運勢は・・・" + fortune_result + "です"


demo = gr.Interface(fn=get_foturne,
                    inputs="text",
                    outputs="text")
demo.launch()
```

```python
import gradio as gr

weather_data = {
    "Tokyo": "Clear sky, 20°C",
    "New York": "Partly cloudy, 15°C",
    "London": "Rainy, 10°C",
    "Paris": "Sunny, 18°C",
}

def get_weather(city):
    return weather_data.get(city, "City not found")

with gr.Blocks() as app:
    with gr.Row():
        city = gr.Textbox(label="City")
        submit_button = gr.Button("Get Weather")
    weather_info = gr.Textbox(label="Weather Information")

    submit_button.click(fn=get_weather, inputs=city, outputs=weather_info)

app.launch()
```

```python
import gradio as gr

# チャット履歴を保存するリスト
chat_history = []

def handle_upload(file):
    content = file.read().decode("utf-8")
    return content

def handle_chat(message):
    global chat_history
    # 新しいメッセージをチャット履歴に追加
    chat_history.append(f"You: {message}")
    # チャット履歴を文字列として結合して返す
    return "\n".join(chat_history)

with gr.Blocks() as app:
    with gr.Tabs():

        with gr.Tab("Chat"):
            with gr.Row():
                chat_message = gr.Textbox(label="Chat")
                chat_button = gr.Button("Send")
            chat_response = gr.Textbox(label="Chat History", elem_id="chat_history", elem_name="chat_history")
            chat_button.click(fn=handle_chat, inputs=chat_message, outputs=chat_response)
        with gr.Tab("File Upload"):
            with gr.Row():
                uploaded_file = gr.File(label="Upload File")
                upload_button = gr.Button("Upload")
            uploaded_file_info = gr.Textbox(label="Uploaded File Content")
            upload_button.click(fn=handle_upload, inputs=uploaded_file, outputs=uploaded_file_info)

app.launch()
```
